```python
# assignment 2 (q2)

# inner join (customer,invoice) tables based on customer_id

data=pd.read_sql('''select customer.customer_id,invoice.total from invoice inner join customer 
                 where customer.customer_id = invoice.customer_id''',con)

# group merged data on customer_id

data_grp=data.groupby(data['customer_id'])

print('Total value of sales for customer: ')

data_grp.sum()
```
